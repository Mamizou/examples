```python
import matplotlib.pyplot as plt
from astropy.visualization import make_lupton_rgb
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
```

```python
g_name = get_pkg_data_filename('./bubble-nebula-0001g300.fit')
r_name = get_pkg_data_filename('./bubble-nebula-0001r300.fit')
B_name = get_pkg_data_filename('./bubble-nebula-0001B300.fit')
g = fits.open(g_name)[0].data
r = fits.open(r_name)[0].data
b = fits.open(B_name)[0].data
```

```python
rgb_default = make_lupton_rgb(r, g, b, Q=10000, filename="bubble_nebula")
plt.imshow(rgb_default, origin='lower')
```
